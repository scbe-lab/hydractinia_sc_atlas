### Detecting and filtering highly variable genes  

```python
%%R
#' plot CV of each gene in a scatter plot with the number of real counts
#' does having lower counts spikes up to higher CVs? Indeed it does (see filter
#' implemented above)

hysim_counts_cv <- merge(
  data.frame(
    id = rownames(hysim_counts),
    counts = as.numeric(rowSums(hysim_counts))
  ),
  data.frame(
    id = rownames(hysim_cpm),
  cv = apply(
    hysim_cpm[,1:38],
    1,
    function(x){
      sd(x)/mean(x)
    }
  )
  ),
  by.x = 1,
  by.y = 1
)
#head(hysim_counts_cv)
#' Normality test
#' Check if the counts and the CV is normally distributed
#'
hysim_counts_cv$counts <- as.numeric(hysim_counts_cv$counts)

plot(density(log(hysim_counts_cv$counts)))
plot(density(na.omit(hysim_counts_cv$cv)))
ks.test(log(hysim_counts_cv$counts),"pnorm")
#' If p-value is less than .05, we reject the null hypothesis.
#' Sample data does not come from a normal distribution.
```

```python
%%R
scatter.smooth(
  log(hysim_counts_cv$counts,10),
  hysim_counts_cv$cv,
  pch=19,
  col=alpha("royalblue",0.1),
  cex=0.7,
  xlab = "log10(sum of counts)",
  ylab = "coefficient of variation",
  main = "Relationship variability / number of counts"
)
```

```python
%%R
#' Loess fitting to generate the highly variable gene set
hysim_loess <- loess(
  hysim_counts_cv$cv ~ log(hysim_counts_cv$counts),
  span= 1,
  family = "symmetric"
)
```

```python
%%R
par(mfrow=c(1,2))
plot(hysim_loess$residuals, col = rgb(0,0,0,0.1))
abline(h = 0,col = "red")
filt <- which(hysim_loess$residuals > 0) #first iteration this was 1
hysim_hvg <- data.frame(id = rownames(hysim_counts_cv$cv[filt]))
scatter.smooth(
  log(hysim_counts_cv$counts),
  hysim_counts_cv$cv,
  pch=19,
  col=alpha("royalblue",0.1),
  cex=0.7,
  xlab = "log(sum of counts)",
  ylab = "coefficient of variation",
  main = "Relationship variability / number of counts"
)
lines( # Visual confirmation that these genes match our criteria or hvg
  log(hysim_counts_cv$counts[filt]),
  hysim_counts_cv$cv[filt],
  pch=19,
  col=alpha("tomato2",0.5),
  cex=0.7,
  type ="p"
)
legend(x = "topright",          # Position
       legend = c("All genes", "Hi-Var genes"),  # Legend texts
       pch = 19,           # Line types
       col = c("royalblue", "tomato2"),           # Line colors
)                 # Line width
par(mfrow=c(1,1))
```

```python
%%R

hysim_ha_clu <- hysim_ctypes_col$col
names(hysim_ha_clu) <- hysim_ctypes_col$cluster

ctypes_rowAnno <-
  rowAnnotation(
    cluster = hysim_ctypes_col$cluster,
    col = list( cluster = hysim_ha_clu ),
    show_legend = F, show_annotation_name = F
    )


clu_ha = HeatmapAnnotation(
  name = "cell types",
  cluster = hysim_ctypes_col$cluster,
  col = list( cluster = hysim_ha_clu )
  )

hysim_cpm_cooc = tidyup(hysim_cpm, highlyvariable = T)
set.seed(4343)# set a fixed seed for reproducibility
h <- c(0.65,0.95) # two levels of average hierarchical clusterings 500 iterations each
clustering_algorithm <- "hclust"
clustering_method <- "average"
cor_method <- "pearson"
p <- 0.15 # percentage of downsampling of the total number of genes
vargenes = rownames(hysim_cpm_cooc)

#Function taken from Levy et al 2021 'treeFromEnsembleClustering' from source
cooc <- treeFromEnsembleClustering(
  x=hysim_cpm_cooc, p=p, h=h,  n = 1000, vargenes = vargenes, bootstrap=F,
  clustering_algorithm=clustering_algorithm, clustering_method=clustering_method,
  cor_method=cor_method
)
```

```python
%%R
cooc_hmap <- Heatmap(
  name="co-occurence",
  cooc$cooccurrence,
  col = colorRamp2(
    c(seq(min(cooc$cooccurrence),
          max(cooc$cooccurrence),
          length=9
    )
    ),
    colors=c(
      c("white",'#ffffe5','#fff7bc','#fee391','#fec44f','#fe9929','#ec7014','#cc4c02','#990000')
    )
  ),
  cluster_rows = as.hclust(cooc$tree),
  cluster_columns = as.hclust(cooc$tree),
  left_annotation = ctypes_rowAnno,
  top_annotation = clu_ha,
  row_names_gp = gpar(fontsize = 8),
  column_names_gp = gpar(fontsize = 8)
)
```

A grouping system for every gene to have an approximation of their peak of gene expression. This will be explored in other scripts

```python
%%R
hysim_genecolor <- data.frame(
  id = rownames(hysim_cpm),
  genecolor = sapply(
    rownames(hysim_cpm),
    function(x) {
      a <- which( rownames(hysim_cpm) == x )
      b <- which( hysim_cpm[a,] == max(hysim_cpm[a,]) ) # which is the highest value of gene i, ie in which ctype is highest
      c <- colnames(
        hysim_cpm[a,][b] # grab the name of the cell type where eigengene i is highest
      )
      d <- hysim_ctypes_col$col[hysim_ctypes_col$cluster == c] #sub("X","",c) # replace some characters
      return(d) # assign in the table
    }
  )
)

head(hysim_genecolor)
```
